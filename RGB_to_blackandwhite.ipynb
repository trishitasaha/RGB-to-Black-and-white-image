```python
import numpy as np 
```

```python
import imageio
```

```python
import scipy.ndimage
```

```python
import matplotlib.pyplot as plt 
```

```python
#Read the image
img='flower.jpeg' 
```

```python
def grayscale(rgb): 
    return np.dot(rgb[..., :3], [0.288, 0.587, 0.114]) 
```

```python
#Dodge the image 
def dodge(front, back): 
    result=front*255/(255-back) 
    result[result>255]=255 
    result[result==255]=255 
    return result.astype('uint8')
```

```python
s=imageio.imread(img) 
```

```python
g=grayscale(s) 
```

```python
i=255-g 
```

```python
#Blur the image 
blur=scipy.ndimage.filters.gaussian_filter(i, sigma=15) 
```

```python
final_img=dodge(blur, g) 
```

```python
#Show the image 
plt.imshow(final_img, cmap='gray') 
```

```python
#Save the image 
plt.imsave('flower_new.jpeg', final_img, cmap='gray', vmin=0, vmax=255) 
```
